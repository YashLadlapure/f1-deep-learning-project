# Notebook 4 – Model Comparison: LSTM vs GNN vs LSTM+GNN
Compare all three deep learning models side-by-side on F1 lap time prediction.

```python
# Cell 1 – Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import sys, os
sys.path.append('../')
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
# Cell 2 – Load saved results (run notebooks 01-03 first)
# Expected files: ../lstm_results.npy, ../gnn_results.npy, ../lstm_gnn_results.npy
results = {}
for name, path in [('LSTM','../lstm_results.npy'),('GNN','../gnn_results.npy'),('LSTM+GNN','../lstm_gnn_results.npy')]:
 if os.path.exists(path):
 results[name] = np.load(path, allow_pickle=True).item()
 print(f'{name}: loaded')
 else:
 print(f'{name}: NOT FOUND – run notebook first')
```

```python
# Cell 3 – Compute metrics table
rows = []
for model_name, res in results.items():
 preds = res['preds']
 targets = res['targets']
 r2 = r2_score(targets, preds)
 mae = mean_absolute_error(targets, preds) / 1000
 rows.append({'Model': model_name, 'R2': round(r2,4), 'MAE (s)': round(mae,4)})
df_metrics = pd.DataFrame(rows)
print(df_metrics.to_string(index=False))
```

```python
# Cell 4 – Bar chart comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
axes[0].bar(df_metrics['Model'], df_metrics['R2'], color=colors)
axes[0].set_title('R² Score Comparison', fontsize=14)
axes[0].set_ylabel('R²')
axes[0].set_ylim(0, 1)
axes[0].grid(axis='y')
axes[1].bar(df_metrics['Model'], df_metrics['MAE (s)'], color=colors)
axes[1].set_title('MAE Comparison (seconds)', fontsize=14)
axes[1].set_ylabel('MAE (s)')
axes[1].grid(axis='y')
plt.tight_layout()
plt.savefig('../model_comparison.png')
plt.show()
```

```python
# Cell 5 – Training loss curves comparison
fig, ax = plt.subplots(figsize=(10, 5))
line_styles = ['-', '--', '-.']
for (model_name, res), ls in zip(results.items(), line_styles):
 ax.plot(res['losses'], label=model_name, linewidth=2, linestyle=ls)
ax.set_title('Training Loss Curves', fontsize=14)
ax.set_xlabel('Epoch')
ax.set_ylabel('MSE Loss')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.savefig('../training_loss_comparison.png')
plt.show()
```

```python
# Cell 6 – Best model summary
best_r2 = df_metrics.loc[df_metrics['R2'].idxmax()]
best_mae = df_metrics.loc[df_metrics['MAE (s)'].idxmin()]
print(f'Best R2 : {best_r2["Model"]} ({best_r2["R2"]:.4f})')
print(f'Best MAE: {best_mae["Model"]} ({best_mae["MAE (s)"]:.4f} s)')
```
